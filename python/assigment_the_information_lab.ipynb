Create and call a python function that : 
- stores a random integer A between 1 and 9
- stores a random integer B between 1 and 9
- multiplies A and B together as C


```python
import random #I import random, a module that generates random intefrates. 

def generate_integers(): # I define the fuction here. Will call it generate integers.
    random_a = random.randint(1, 9) # first variable, random_a will return a random number from 1 to 9.
    random_b = random.randint(1, 9) #same as before, in this case the variable is called random_b.
    random_c = random_a * random_b #third part of the excersice, multiply 'a' and 'b' to get 'c'.

    print(f"Random integer A: {random_a}") #Once I run this code, it will be printed with the disered format.
    print(f"Random integer B: {random_b}")
    print(f"Random integer C,  result of Random_a * Random_b: {random_c}")
    return random_a, random_b, random_c # so we can see what numbers are they multiplying.

generate_integers(); #calling the function to see the outcome. ';' so I only get the 'print' part.
```
